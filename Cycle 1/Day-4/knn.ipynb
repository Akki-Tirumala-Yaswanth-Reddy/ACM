```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Day-2/after_day2.csv')
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df[[col for col in df.columns if col != "BurnoutRisk"]]
y = df["BurnoutRisk"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=8)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

y_pred_tr = tree.predict(X_test)
accuracy_score(y_test, y_pred_tr)
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forrest = RandomForestClassifier()
random_forrest.fit(X_train, y_train)
```

```python

y_pred_rf = random_forrest.predict(X_test)
accuracy_score(y_test, y_pred_rf)
```

```python
features = random_forrest.feature_importances_
features = pd.Series(data=features, index= X.columns)
features.sort_values(ascending=False)
```

```python
top3 = features.sort_values(ascending=False).head(3).index
top3
```

```python
X = df[top3]
y = df["BurnoutRisk"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kclassi = KNeighborsClassifier(n_neighbors=2)
kclassi.fit(X_train, y_train)

y_pred_knn = kclassi.predict(X_test)

print("Accuracy of KNN with only 3 features is: ", accuracy_score(y_test, y_pred_knn))
```
