Handling ordinal values

```python
df['Workout Intensity'] = df['Workout Intensity'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Mood Before Workout'] = df['Mood Before Workout'].map({ 'Stressed': 0,'Tired': 1, 'Neutral': 2, 'Happy': 3})
df['Mood After Workout'] = df['Mood After Workout'].map({ 'Fatigued': 0,'Neutral': 2, 'Energized': 3})
```

```python
df.head()
```

Change in mood will be used to create our target column and every value below 0 will be considered as droping out.

```python
df['change_in_mood'] = df['Mood Before Workout'] - df['Mood After Workout']
df['change_in_mood'].value_counts()
```

```python
df['target'] = df['change_in_mood'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df['target'].head()
```

Handling nominal values

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['Workout Type', 'Gender']])
encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out(['Workout Type', 'Gender']))
df = pd.concat([df, encoded_df], axis=1)
df.drop(['Workout Type', "Gender"], axis=1, inplace=True)
```

```python
df.head()
```

Normalizing data

```python
from sklearn.preprocessing import Normalizer

scaler = Normalizer()
df[numerical_col] = scaler.fit_transform(df[numerical_col])
df.head()
```

```python
import torch

X = torch.tensor(df.drop('target', axis=1).values, dtype=torch.float32)
y = torch.tensor(df['target'].values, dtype=torch.long)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=8)
```

```python
from torch.utils.data import Dataset, DataLoader

class WorkoutDataset(Dataset):
  def __init__(self, X, y):
    self.X = X
    self.y = y

  def __len__(self):
    return len(self.y)

  def __getitem__(self, idx):
    return self.X[idx], self.y[idx]
```

```python
train_dataset = WorkoutDataset(X_train, y_train)
test_dataset = WorkoutDataset(X_test, y_test)
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size = 32)
```

```python
from torch import nn

class NN(nn.Module):
  def __init__(self):
    super(NN, self).__init__()
    self.layer1 = nn.Linear(28, 7)
    self.layer2 = nn.Linear(7, 1)
    self.layer3 = nn.Sigmoid()


  def forward(self, x):
    x = torch.relu(self.layer1(x))
    x = torch.relu(self.layer2(x))
    x = self.layer3(x)
    return x
```

```python
model = NN()
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
import torch.nn.functional as F

for i in range(5):
  epoch_loss = 0
  for X, y in train_dataloader:
    output = model(X)
    loss = loss_fn(output, y.unsqueeze(1).float())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    epoch_loss += loss.item()
  print(f"Epoch {i+1}, Loss: {epoch_loss/len(train_dataloader):.4f}")
```

```python

y_true = []
y_pred = []

with torch.no_grad():
  for X, y in test_dataloader:
    output = model(X)
    predictions = torch.where(output > 0.5, 1, 0)
    y_true.extend(y.tolist())
    y_pred.extend(predictions.tolist())

y_true = torch.tensor(y_true).cpu().numpy()
y_pred = torch.tensor(y_pred).cpu().numpy()
```

```python
for k in range(10):
  print(y_pred[k], y_true[k])
```

```python
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

print('Accuracy: ', accuracy_score(y_true, y_pred))
print('Precision score: ', precision_score(y_true, y_pred))
print('F1 score: ', f1_score(y_true, y_pred))
print('Recall socre: ', recall_score(y_true, y_pred))
```
